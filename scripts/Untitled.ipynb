# Write functions

```python
def smilesList(path):
    '''
    Generate list of SMILES from path
    '''
    with open(path) as f:
        data = [line.strip() for line in f]
    return data
```

```python
def getNAtoms(smiles):
    mol = Chem.MolFromSmiles(smiles)
    try:
        Chem.AddHs(mol)
        return None
    except:
        print smiles
        return smiles
```
